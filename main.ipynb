```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carregar a imagem
imagem = cv2.imread('ifg.jpg')
```

```python
# Converter para tons de cinza
imagem_cinza = imagem
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

imagem_cinza_rgb = cv2.cvtColor(imagem_cinza, cv2.COLOR_BGR2RGB)

# Exibir a imagem usando matplotlib
plt.figure(figsize=(10, 8))
plt.imshow(imagem_cinza_rgb)
plt.axis('off')  
plt.title('imagem_cinza_rgb')
plt.show()
```

```python
# Aplicar filtro Gaussiano para suavizar a imagem
imagem_suavizada = cv2.GaussianBlur(imagem_cinza, (3, 3), 0)

imagem_suavizada_rgb = cv2.cvtColor(imagem_suavizada, cv2.COLOR_BGR2RGB)

# Exibir a imagem usando matplotlib
plt.figure(figsize=(10, 8))
plt.imshow(imagem_suavizada_rgb)
plt.axis('off')  
plt.title('imagem_suavizada_rgb')
plt.show()
```

```python
# Aplicar limiarizao (threshold) para binarizar a imagem
_, imagem_binaria = cv2.threshold(imagem_suavizada, 127, 255, cv2.THRESH_BINARY_INV)

imagem_binaria_rgb = cv2.cvtColor(imagem_binaria, cv2.COLOR_BGR2RGB)

# Exibir a imagem usando matplotlib
plt.figure(figsize=(10, 8))
plt.imshow(imagem_binaria_rgb)
plt.axis('off')  
plt.title('imagem_binaria_rgb')
plt.show()
```

```python
# Encontrar contornos na imagem binria
contornos, _ = cv2.findContours(imagem_binaria, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# Desenhar contornos na imagem original
cv2.drawContours(imagem, contornos, -1, (255, 0, 255), 2)
```

```python
# Exibir a imagem com os contornos desenhados
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

# Exibir a imagem usando matplotlib
plt.figure(figsize=(10, 8))
plt.imshow(imagem_binaria_rgb)
plt.axis('off')  
plt.title('imagem_binaria_rgb')
plt.show()
```

```python
# Contar o nmero de objetos detectados
numero_de_objetos = len(contornos)
print(f"Total de objetos detectados: {numero_de_objetos}")
```

```python
cv2.drawContours(imagem, contornos, -1, (255, 0, 255), 2)

imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 8))
plt.imshow(imagem)
plt.axis('off')  
plt.title('Objetos detectados')
plt.show()
```
